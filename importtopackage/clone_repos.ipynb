```python
import sqlite3
import json

with open("shared.json") as f:
    dois = json.load(f)

# Dataset of a Study of Computational reproducibility of Jupyter notebooks from biomedical publications
# https://zenodo.org/records/8226725
con = sqlite3.connect("../../hackathon/computational-reproducibility-pmc/computational-reproducibility-pmc/analyses/db.sqlite")
cur = con.cursor()
```

```python
import os
from operator import or_
from functools import reduce


class Hashabledict(dict):
    def __hash__(self):
        return hash(frozenset(self))

ok = {}
fail = {}

imports_in_notebooks_by_doi = {}
for doi in dois:
    res = cur.execute("""
        select r.repository, r.`commit`
        from repositories r
        inner join article a on (a.id = r.article_id)
        where a.doi = ?
    """, (doi,))
    for row in res.fetchall():
        try:
            %cd ~/projects/hackathon/projects
            os.makedirs(row[0], exist_ok=True)
            %cd {row[0]}
            !git init
            !git remote add origin https://github.com/{row[0]}.git
            !git fetch --depth 1 origin {row[1]}
            !git checkout FETCH_HEAD
            ok[doi] = row
        except:
            fail[doi] = row
```
